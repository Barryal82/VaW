```python

# Specify the columns of interest
columns_of_interest = ['Age', 'Gender', 'Current_Income_Status', 'Respondent_Status']

# Loop over each column of interest
for column in columns_of_interest:
    # Calculate correlation coefficients between the current column and all other columns
    column_correlation = df.corrwith(df[column])
    
    # Convert correlation coefficients into a DataFrame
    correlation_df = pd.DataFrame(column_correlation, columns=['Correlation'])
    
    # Check if the DataFrame is empty or contains only NaN values
    if correlation_df.empty or correlation_df['Correlation'].isnull().all():
        print(f"No valid data to plot heatmap for column '{column}'.")
    else:
        # Generate heatmap if the DataFrame is not empty
        plt.figure(figsize=(10, 1))
        sns.heatmap(correlation_df.transpose(), annot=True, cmap='coolwarm', fmt=".1f")
        plt.title(f'Correlation Matrix with {column}')
        plt.show()
```

```python
from scipy.stats import spearmanr

# Specify the columns of interest
columns_of_interest = ['Age', 'Gender', 'Current_Income_Status', 'Respondent_Status']

# Initialize an empty dictionary to store correlation results
correlation_results = {}

# Iterate over each column of interest
for column in columns_of_interest:
    # Initialize an empty dictionary to store correlations for the current column
    column_correlations = {}
    # Iterate over each column in the DataFrame
    for other_column in df.columns:
        # Skip the column of interest itself
        if other_column == column:
            continue
        # Calculate Spearman correlation coefficient and p-value
        correlation_coefficient, p_value = spearmanr(df[column], df[other_column])
        # Store the results in the dictionary
        column_correlations[other_column] = {'Spearman correlation coefficient': correlation_coefficient,
                                             'p-value': p_value}
    # Store the correlations for the current column in the overall dictionary
    correlation_results[column] = column_correlations

# Print the results
for column, column_correlations in correlation_results.items():
    print(f"Correlations for '{column}':")
    for other_column, result in column_correlations.items():
        print(f"  '{other_column}': Spearman correlation coefficient = {result['Spearman correlation coefficient']}, p-value = {result['p-value']}")
```

```python
import plotly.graph_objs as go

# Group the data by 'Age' and 'Gender' and count the number of occurrences
age_gender_counts = df.groupby(['Age', 'Gender']).size().reset_index(name='Count')

# Define colors for each gender
gender_colors = {1: '#1f77b4', 2: '#ff7f0e'}  # Blue for Male, Orange for Female

# Create a trace for each gender
traces = []
for gender, color in gender_colors.items():
    gender_data = age_gender_counts[age_gender_counts['Gender'] == gender]
    trace = go.Funnel(
        y=gender_data['Age'],
        x=gender_data['Count'],
        name='Male' if gender == 1 else 'Female',
        marker={'color': color}
    )
    traces.append(trace)

# Define layout for the funnel chart
layout = go.Layout(
    title='Age Groups Disaggregated by Gender',
    funnelmode="stack",
    hovermode='closest',
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
    xaxis=dict(title='Count'),
    yaxis=dict(title='Age Group')
)

# Create the figure and plot the funnel chart
fig = go.Figure(data=traces, layout=layout)
fig.show()
```
